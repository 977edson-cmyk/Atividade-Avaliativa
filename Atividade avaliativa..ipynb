```python
1) Cálculo do peso ideal:
# Entrada de dados
altura = float(input("Digite sua altura (em metros): "))
sexo = input("Digite seu sexo (M para masculino, F para feminino): ").upper()

# Cálculo do peso ideal
if sexo == 'M':
    peso_ideal = (72.7 * altura) - 58
elif sexo == 'F':
    peso_ideal = (62.1 * altura) - 44.7
else:
    peso_ideal = None
    print("Sexo inválido.")

# Exibição do resultado
if peso_ideal is not None:
    print(f"O seu peso ideal é: {peso_ideal:.2f} kg")

2) Cálculo da idade e verificação para votar e tirar habilitação:
# Entrada de dados
ano_nascimento = int(input("Digite seu ano de nascimento: "))
ano_atual = 2025  # Considerando o ano atual como 2025

# Cálculo da idade
idade = ano_atual - ano_nascimento

# Verificação
if idade >= 16:
    pode_votar = "Você pode votar."
else:
    pode_votar = "Você ainda não pode votar."

if idade >= 18:
    pode_habilitacao = "Você pode tirar a Carteira de Habilitação."
else:
    pode_habilitacao = "Você ainda não pode tirar a Carteira de Habilitação."

# Exibição do resultado
print(f"Sua idade é: {idade} anos.")
print(pode_votar)
print(pode_habilitacao)

3) Classificação de um produto com base no código:
# Entrada de dados
codigo_produto = int(input("Digite o código do produto: "))

# Classificação do produto
if codigo_produto == 1:
    classificacao = "Alimento não-perecível"
elif codigo_produto in [2, 3, 4]:
    classificacao = "Alimento perecível"
elif codigo_produto in [5, 6]:
    classificacao = "Vestuário"
elif codigo_produto == 7:
    classificacao = "Higiene pessoal"
elif 8 <= codigo_produto <= 15:
    classificacao = "Limpeza e utensílios domésticos"
else:
    classificacao = "Inválido"

# Exibição do resultado
print(f"A classificação do produto é: {classificacao}")

4) Classificação de nadador por idade:
# Entrada de dados
idade_nadador = int(input("Digite a idade do nadador: "))

# Classificação do nadador
if 5 <= idade_nadador <= 7:
    categoria = "Infantil A"
elif 8 <= idade_nadador <= 10:
    categoria = "Infantil B"
elif 11 <= idade_nadador <= 13:
    categoria = "Juvenil A"
elif 14 <= idade_nadador <= 17:
    categoria = "Juvenil B"
elif idade_nadador >= 18:
    categoria = "Adulto"
else:
    categoria = "Idade inválida para nadador."

# Exibição do resultado
print(f"A categoria do nadador é: {categoria}")

5) Cálculo do preço a pagar com base no código de pagamento:
# Entrada de dados
preco_produto = float(input("Digite o preço do produto: R$ "))
condicao_pagamento = int(input("Digite o código da condição de pagamento (1, 2, 3, ou 4): "))

# Cálculo do preço com base na condição de pagamento
if condicao_pagamento == 1:
    total_a_pagar = preco_produto * 0.90  # 10% de desconto
    print(f"Você pagará R${total_a_pagar:.2f} com 10% de desconto.")
elif condicao_pagamento == 2:
    total_a_pagar = preco_produto  # Preço integral
    print(f"Você pagará R${total_a_pagar:.2f} no preço integral.")
elif condicao_pagamento == 3:
    total_a_pagar = preco_produto * 1.10  # 10% de juros
    print(f"Você pagará R${total_a_pagar:.2f} com 10% de juros.")
elif condicao_pagamento == 4:
    total_a_pagar = preco_produto * 1.20  # 20% de juros
    print(f"Você pagará R${total_a_pagar:.2f} com 20% de juros.")
else:
    print("Código de pagamento inválido.")
```
